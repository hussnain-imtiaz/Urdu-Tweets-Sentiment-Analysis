### Vectorizing the Text 

```python
#TF-IDF vectorizer
tf=TfidfVectorizer()
```

```python
X_weights=tf.fit_transform(df["Text"])
X_weights.shape
```

```python
X_weights
```

```python
### having a look over vectorized dataframe
X_weights_df=pd.DataFrame(X_weights.toarray(),columns=tf.get_feature_names())
```

```python
X_weights_df.head()
```

```python
# Feel free to use different ratios to split the data.
X_train, X_test, y_train, y_test = train_test_split(X_weights_df,df.Sentiment, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python

np.random.seed(42)
# train model

#Ensemble Method
RFC = RandomForestClassifier().fit(X_train, y_train)

#Linear Model
LR=LogisticRegression().fit(X_train, y_train)

#SVM Classifier
SV=SVC().fit(X_train, y_train)

#Naive Bayes
NV=MultinomialNB().fit(X_train, y_train)

#SGD Classifier

SGC=SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3).fit(X_train, y_train)
```

```python
#model testing
def test_model(clf, X_test,y_test):
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, pos_label="P", average='binary')
    print(f'Classifier: {clf} \n')
    print('Accuracy: ', acc)
    print('Precision: ', pre)
    print('Recall: ', rec)
    print('f1-score: ', f1)
    print(" \n")
```

```python
classifiers=pd.Series([RFC,LR,SV,NV,SGC])
```

```python
classifiers.apply(lambda x : test_model(x,X_test,y_test))
```
